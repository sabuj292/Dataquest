```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

**Comparing across all degrees**

```python

c_a =(99/255,210/255,190/255)
c_b = (150/255,230/255,70/255) 
# Generate line charts for both male and female for every degree in the stem_cats
fig = plt.figure(figsize=(18,20))


for sp in range(0,18,3):
    ind = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[ind]],c=c_a,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[ind]], c= c_b,label='Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top='off',bottom='off',left='off',right='off')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    if ind == 0:
        ax.text(2005,10,'Men')
        ax.text(2002,90,'Women')
    elif ind == 5:
        ax.text(2005,100,'Men')
        ax.text(2002,10,'Women')
    ax.set_title(stem_cats[ind])



#Generating line charts for Lib_arts_cats

for sp in range(1,16,3):
    ind = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[ind]],c=c_a,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[ind]], c= c_b,label='Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top='off',bottom='off',left='off',right='off')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    if ind == 0:
        ax.text(2005,20,'Men')
        ax.text(2002,80,'Women')
    
    ax.set_title(lib_arts_cats[ind])
    
    
#Generating line charts for other_cats
for sp in range(2,20,3):
    ind= int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[ind]],c=c_a,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[ind]], c= c_b,label='Men',linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.tick_params(top='off',bottom='off',left='off',right='off')
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    if ind == 0:
        ax.text(2005,10,'Men')
        ax.text(2002,90,'Women')
    elif ind == 5:
        ax.text(2005,70,'Men')
        ax.text(2002,30,'Women')
    ax.set_title(other_cats[ind])
    



plt.show()
    
    
```
